```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = 'PS2'

vals = []

arrx = []
for i in os.listdir(path):
    arrx.append(np.load(f'{path}/{i}'))
    print(i.split('.')[0].split('-')[-1])
    vals.append(i.split('.')[0].split('-')[-1])
```

```python
from tqdm import tqdm


fig = plt.figure(figsize=(20, 20)) 
figsL = ['LL','LP',"RP","RR"]
for i in tqdm(range(1,21)):
    fig.add_subplot(4,5,i)
    
    if i%5 == 4:
        # print((i-1)//5)
        plt.title(f'{vals[0]} - {figsL[(i-1)//5]}')
        # print((i-1)//5)
        plt.imshow(arrx[0][:,:,(i-1)//5])
        # plt.plot(range(x1),(EEG1[(i-1)//2]))
    

    if i%5 == 2:
        # print((i-1)//5)
        plt.title(f'{vals[1]} - {figsL[(i-1)//5]}')
        print((i-1)//5)
        plt.imshow(arrx[1][:,:,(i-1)//5])
        # plt.plot(range(x1),(EEG1[(i-1)//2]))

    if i%5 == 1:
        # print((i-1)//5)
        plt.title(f'{vals[2]} - {figsL[(i-1)//5]}')
        print((i-1)//5)
        plt.imshow(arrx[2][:,:,(i-1)//5])
        # plt.plot(range(x1),(EEG1[(i-1)//2]))

    if i%5 == 0:
        # print((i-1)//5)
        plt.title(f'{vals[3]} - {figsL[(i-1)//5]}')
        print((i-1)//5)
        plt.imshow(arrx[3][:,:,(i-1)//5])
        # plt.plot(range(x1),(EEG1[(i-1)//2]))
    if i%5 == 3:
        # print((i-1)//5)
        plt.title(f'{vals[4]} - {figsL[(i-1)//5]}')
        print((i-1)//5)
        plt.imshow(arrx[4][:,:,(i-1)//5])
        # plt.plot(range(x1),(EEG1[(i-1)//2]))

    # if i%6 == 0:
    #     # print((i-1)//5)
    #     plt.title(f'db8 - {figsL[(i-1)//6]}')
    #     print((i-1)//6)
    #     plt.imshow(arrx[4][:,:,(i-1)//6])
    #     # plt.plot(range(x1),(EEG1[(i-1)//2]))
    fig.suptitle('Processed Spectrograms - Using Superlets',fontsize=40)
    
```
